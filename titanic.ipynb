```python
import numpy as np  # 계산
import pandas as pd # 분석
import matplotlib.pyplot as plt # 그림
import seaborn as sns # 그림

%matplotlib inline
```

```python
raw_data = pd.read_excel('titanic.xls')
```

```python
raw_data.info()

# pclass 선실등급
# survived 생존 여부
# age 나이
# sibsp 함께 탑승한 형제자매 또는 배우자의 수
# parch 함께 탑승한 부모나 자녀의 수
# fare 요금
```

```python
raw_data.head()
```

```python
raw_data.describe()
```

```python
# 생존자 수 분석

f, ax = plt.subplots(1, 2, figsize = (12,6))
raw_data['survived'].value_counts().plot.pie(explode = [1, 0.1],
                                             autopct = '%1.2f%%', ax=ax[0])
sns.countplot('survived', data=raw_data, ax=ax[1])
plt.show()
```

```python
# 연령 분석

raw_data['age'].hist(bins=20, figsize=(18,8), grid=False)
```

```python
# 등급별 분석

raw_data.groupby('pclass').mean()
```

```python
# 상관계수 분석 함수 corr()
raw_data.corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(raw_data.corr(), linewidths=0.01, square=True, 
            annot = True, cmap=plt.cm.viridis, linecolor='white')  # 값의 상관관계에 따라 색을 반환
```

```python
# 나이 구간별 성별 생존율 분석

# [0, 3, 7, 15, 30, 60, 100]

raw_data['age_cat'] = pd.cut(raw_data['age'],
                             bins = [0, 3, 7, 15, 30, 60, 100],
                             labels=['baby', 'children', 'teenager', 'young', 'adult', 'old'])

raw_data.info()
```

```python
raw_data.groupby('age_cat').mean()
```

```python
# Quiz 등급별 생존률, 나이 카테고리별 생존률, 성별 생존률
# 3개를 옆으로 한줄로 표현하기
```

```python
f, ax = plt.subplots(1, 3, figsize=(12, 6))

sns.barplot('pclass', 'survived', data = raw_data, ax = ax[0])
sns.barplot('age_cat', 'survived', data = raw_data, ax = ax[1])
sns.barplot('sex', 'survived', data = raw_data, ax = ax[2])

plt.show()
```
