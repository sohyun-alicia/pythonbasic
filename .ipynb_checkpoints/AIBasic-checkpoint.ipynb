# 모델 생성

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# X는 대문자, y는 소문자(관용적 쓰임)
X = iris.data[:,2:]    
# petal에 width와 length를 사용하기 위해 X변수에 모든 행에서 2번째 칼럼 데이터부터 사용

y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 2, random_state = 13)
tree_clf.fit(X,y)
```

# 시각화 도구 Graphviz 설치

```python
!pip install graphviz
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(
    tree_clf,
    out_file = "iris_tree.dot",
    feature_names = ['petal_length','petal_width'],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)
```

```python
import graphviz
with open("iris_tree.dot") as f:
    dot_graph = f.read()
    
dot = graphviz.Source(dot_graph)
dot.format = 'png'
dot.render(filename='iris_tree', directory='images/decision_trees', cleanup=True)
dot
```
